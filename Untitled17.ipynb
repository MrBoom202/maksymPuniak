1. Змінна а набуває значення «Україна», а змінна b — «Франція». За допомогою
інструкції input() уводиться довільне ціле число. Розробіть програму без методів, у
якій створюються два об’єкти. Якщо уведене число більше 5, виводиться значення з
використанням першого об’єкта, інакше — з використанням другого об’єкта.

```python
class Country:
    def __init__(self, name):
        self.name = name

a = Country("Україна")
b = Country("Франція")

number = int(input("Введіть ціле число: "))

if number > 5:
    print(a.name)
else:
    print(b.name)
```

3. Розробіть програму з використанням двох методів. В одному з них обчислюється
об’єм кулі радіусом r, а в другому — площа трикутника зі сторонами a, b, c.

```python
import math

class Geometry:
    def sphere_volume(self, r):
        return (4/3) * math.pi * r**3

    def triangle_area(self, a, b, c):
        if a + b > c and a + c > b and b + c > a:
            p = (a + b + c) / 2
            return math.sqrt(p * (p - a) * (p - b) * (p - c))
        else:
            return None

g = Geometry()

r = float(input("Введіть радіус: "))
print("Обʼєм кулі =", g.sphere_volume(r))

a = float(input("Введіть сторону a: "))
b = float(input("Введіть сторону b: "))
c = float(input("Введіть сторону c: "))

area = g.triangle_area(a, b, c)

if area is None:
    print("Трикутник не існує")
else:
    print("Площа трикутника =", area)

```

1. Конструктор має такі значення параметрів: self, a1 = 3, a2 = 4. Створіть програму
множення значень параметрів конструктора, за умови, що він викликається — з
першого об’єкта без аргументів, з другого об’єкта — з аргументом 5 і 6, а  з третього
— з аргументом 7.

```python
class Multiply:
    def __init__(self, a1=3, a2=4):
        self.a1 = a1
        self.a2 = a2

    def result(self):
        return self.a1 * self.a2

obj1 = Multiply()
obj2 = Multiply(5, 6)
obj3 = Multiply(7)

print("Результат 1 =", obj1.result())
print("Результат 2 =", obj2.result())
print("Результат 3 =", obj3.result())
```

2. Розробіть програму, у якій конструктор має параметри self, a1 і a2. Він викликається з
двох об’єктів, аргументами яких є два цілих числа. Якщо значення першого аргумента
більше другого, ці числа множаться, інакше — значення першого ділиться на друге.

```python
class Numbers:
    def __init__(self, a1=3, a2=4):
        self.a1 = a1
        self.a2 = a2

    def calculate(self):
      if self.a1 > self.a2:
        return self.a1 * self.a2
      else:
        return self.a1 / self.a2

obj1 = Numbers(int(input("Введіть перше число для об'єкта 1: ")),
               int(input("Введіть друге число для об'єкта 1: ")))

obj2 = Numbers(int(input("Введіть перше число для об'єкта 2: ")),
               int(input("Введіть друге число для об'єкта 2: ")))

print("Результат 1 =", obj1.calculate())
print("Результат 2 =", obj2.calculate())
```

3. Конструктор має параметри зі значеннями: self, a1 = &quot;Операційна&quot; a2 = &quot;система&quot;. За
допомогою інструкції input() уводиться слово. Якщо це слово Linux, виводиться
повідомлення Операційна система Linux, інакше — повідомлення Операційна система
Windows.

```python
class OS:
    def __init__(self, a1="Операційна", a2="система"):
        self.a1 = a1
        self.a2 = a2

    def show(self, name):
        if name == "Linux":
            print(self.a1, self.a2, "Linux")
        else:
            print(self.a1, self.a2, "Windows")

obj = OS()

word = input("Введіть слово: ")
obj.show(word)
```
