```python
from matplotlib import pyplot as plt
from matplotlib_venn import venn3

A = {1, 2, 3, 4, 5, 6, 7}
B = {5, 6, 7, 8, 9, 10}
C = {1, 2, 3, 8, 9, 10}

U = A | B | C
A_inter_C = A & C
B_complement = U - B
result = A_inter_C - B_complement

print("A ∩ C =", A_inter_C)
print("B_ =", B_complement)
if result == set():
    print("3) (A ∩ C) \\ (B_) = Порожня множина")
else:
    print("3) (A ∩ C) \\ (B_) =", result)

v = venn3([A, B, C], ('A', 'B', 'C'))
v.get_label_by_id('111').set_text('\n'.join(map(str, result)) if result else '∅')
plt.title("Діаграма для (A ∩ C) \\ (B^c)")
plt.show()
```
